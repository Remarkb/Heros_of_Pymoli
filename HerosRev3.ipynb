## Heros of Pymoli Observations

- A majority of "Heros of Pymoli" players are male, making up slightly over 84% of total players.  The percentage of female players is smaller, but not insignificant at 14%.
- While males accounted for a much greater share of total purchase value, females had a greater average total purchase per person at $4.47$ vs the males $4.07$.
- Nearly 77% of the players fall within a 15 year age range between 15 years of age and 29 years of age.
- Not surprisingly, that same age range of 15 - 29 accounts for nearly 77% of the total purchase value as well.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)

purchase_data_df.head()
```

## Player Count

* Display the total number of players


```python
# Find unique players
player_uniques_df = purchase_data_df.loc[:, ["SN"]]
player_uniques_df = player_uniques_df.drop_duplicates()

# Count the unique players
player_count_df = player_uniques_df.SN.count()

pd.DataFrame({"Total Players": [player_count_df]})
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Number of Unique Items, Average Price, Number of Purchases, Total Revenue
average_price = purchase_data_df["Price"].mean()
total_revenue = purchase_data_df["Price"].sum()
total_purchases = purchase_data_df["Price"].count()

Unique_Items = purchase_data_df.loc[:, ["Item ID"]]
Unique_Items = Unique_Items.drop_duplicates()
Unique_Items_count = Unique_Items["Item ID"].count()

purchase_analysis = pd.DataFrame({"Average Price": [average_price], "Total Revenue": [total_revenue], "Total Purchases": [total_purchases], "Unique Items Purchased": [Unique_Items_count]})
purchase_analysis["Average Price"] = purchase_analysis["Average Price"].astype(float).map("${:,.2f}".format)
purchase_analysis["Total Revenue"] = purchase_analysis["Total Revenue"].astype(float).map("${:,.2f}".format)
#print(average_price, total_revenue, total_purchases, Unique_Items_count)
purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_demos_df = pd.DataFrame(purchase_data_df.drop_duplicates(["SN"]))
gender_demos_group = gender_demos_df.groupby(["Gender"])
gender_breakdown = pd.DataFrame({"Total Count":gender_demos_group["Gender"].count(),"Percentage of Players":gender_demos_group["Gender"].count()/player_count_df*100})

gender_breakdown["Percentage of Players"] = gender_breakdown["Percentage of Players"].astype(float).map("{:,.2f}%".format)

gender_breakdown.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#gen_purch_count
#avg_purch_price
#total_purch_value
#avg_total_purchase

gender_analysis = purchase_data_df.groupby(["Gender"])

gen_purch_count = gender_analysis["SN"].count()
avg_purch_price = gender_analysis["Price"].mean()
total_purch_value = gender_analysis["Price"].sum()
avg_tot_purch = total_purch_value / gender_breakdown["Total Count"]

gender_purch_analysis = pd.DataFrame({"Purchase Count": gen_purch_count, "Average Purchase Price": avg_purch_price, "Total Purchase Value": total_purch_value, "Avg Total Purchase per Person":avg_tot_purch})
gender_purch_analysis["Average Purchase Price"] = gender_purch_analysis["Average Purchase Price"].astype(float).map("${:,.2f}".format)
gender_purch_analysis["Total Purchase Value"] = gender_purch_analysis["Total Purchase Value"].astype(float).map("${:,.2f}".format)
gender_purch_analysis["Avg Total Purchase per Person"] = gender_purch_analysis["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

gender_purch_analysis
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 50]
bin_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

age_group = purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"], bins, labels=bin_labels)
age_group_bin = purchase_data_df.groupby(["Age Group"])


age_purch_count = age_group_bin["SN"].count()
ageavg_purch_price = age_group_bin["Price"].mean()
agetotal_purch_value = age_group_bin["Price"].sum()
ageavg_tot_purch = agetotal_purch_value / age_breakdown["Total Count"]

age_purch_analysis = pd.DataFrame({"Purchase Count": age_purch_count, "Average Purchase Price": ageavg_purch_price, "Total Purchase Value": agetotal_purch_value, "Avg Total Purchase per Person": ageavg_tot_purch})
age_purch_analysis["Average Purchase Price"] = age_purch_analysis["Average Purchase Price"].astype(float).map("${:,.2f}".format)
age_purch_analysis["Total Purchase Value"] = age_purch_analysis["Total Purchase Value"].astype(float).map("${:,.2f}".format)
age_purch_analysis["Avg Total Purchase per Person"] = age_purch_analysis["Avg Total Purchase per Person"].astype(float).map("${:,.2f}".format)

age_purch_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#big_spenders = purchase_data_df("SN")
big_spenders = pd.DataFrame(purchase_data_df)
big_spenders_grp = big_spenders.groupby("SN")

top_purch_count = big_spenders_grp["SN"].count()
topavg_purch_price = big_spenders_grp["Price"].mean()
toptotal_purch_value = big_spenders_grp["Price"].sum()

top_purch_analysis = pd.DataFrame({"Purchase Count": top_purch_count, "Average Purchase Price": topavg_purch_price, "Total Purchase Value": toptotal_purch_value})
top_purch_analysis.sort_values("Total Purchase Value", ascending=False).nlargest(5, "Total Purchase Value")
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items = purchase_data_df.loc[:, ["Item ID", "Item Name", "Price"]]
popular_items_grp = popular_items.groupby(["Item ID", "Item Name"])

pop_purch_count = popular_items_grp["Item ID"].count()
pop_price = popular_items_grp["Price"].first()
poptotal_purch_value = popular_items_grp["Price"].sum()

pop_purch_analysis = pd.DataFrame({"Purchase Count": pop_purch_count, "Item Price": pop_price, "Total Purchase Value": poptotal_purch_value})

pop_purch_analysis["Item Price"] = pop_purch_analysis["Item Price"].astype(float).map("${:,.2f}".format)
pop_purch_analysis["Total Purchase Value"] = pop_purch_analysis["Total Purchase Value"].astype(float).map("${:,.2f}".format)

pop_purch_analysis.sort_values("Purchase Count", ascending=False).nlargest(5, "Purchase Count")
```
